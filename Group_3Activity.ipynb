# Countplot of Outcome

```python
sns.countplot(data['Outcome'])
```

# Effect of Age on Outcome

```python
sns.boxplot(x='Outcome',y='Age',data=data)
plt.title('Age vs Outcome')
```

As age increases chance of getting diabetes also increases

# Chance of getting diabetes

```python
sns.boxplot(x='Outcome',y='DiabetesPedigreeFunction',data=data)
plt.title('Outcome vs DiabetesPedigreeFunction ')
```

# Relationship with Insulin and Glucose

```python
sns.scatterplot(data['Insulin'],data['Glucose'],hue=data['Outcome'])
plt.title('Plot of Insulin vs Glucose',fontsize=16)
plt.figure(figsize=(8,8))
plt.show()
```

# Heatmap on Correlation

```python
sns.heatmap(data.corr(),annot=True)
plt.xticks(rotation=90)
```
