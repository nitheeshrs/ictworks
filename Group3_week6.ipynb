```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel('Dataset_w6.xlsx')
```

```python
data
```

# No of Countries

```python
country=data['country'].count()
print('Number of countries:',country)
```

# Visualization

```python
plt.figure(figsize=(10,5))
x = data['Region']
y = data['Population in thousands (2017)']
plt.xlabel('Region')
plt.ylabel('Population in thousands')
plt.bar(x,y)
plt.xticks(rotation=90);
```

The population of EasternAsia region is higher than others.Also the population of Polynesia and Micronesia is significantly
small compared to others.

```python
plt.figure(figsize=(10,5))
x = data['Region']
y = data['GDP per capita (current US$)']
plt.xlabel('Region')
plt.ylabel('GDP per capita (current US$)')
plt.bar(x,y,color= 'red')
plt.xticks(rotation=90);
```

The highest GDP per capita is for WesternEurope and the lowest is for WesternAfrica.

# No of Countries with respect to region 

```python
data.groupby('Region')["country"].count()
```

# Average on GDP and Intenational trade

```python
plt.figure(figsize=(10,5))
data1=data.groupby('Region')['GDP per capita (current US$)'].mean().plot(kind='bar')
plt.title("Average GDP Per Capita")
plt.show()
```

On average NorthAmerica has the highest GDP per capita.

```python
plt.figure(figsize=(10,5))
data1=data.groupby('Region')['International trade: Imports (million US$)'].mean().plot(kind='bar')
plt.title("Average Inernational trade:Imports")
plt.show()
```

NorthAmerica comes top on average International trade: Imports 

# More Consistent Region with respect to GDP

```python
data2=data.groupby('Region')['GDP per capita (current US$)'].std()
data2
```

```python
data2.nsmallest(1)
```

WesternAfrica has less standard deviation.Hence it is more consistent with respect to GDP
