# Check whether all the zones generate the same amount of sales.

H0 : All the zones generate same amount of sales
    
H1 : All the zones generate different amount of sales    

```python
fvalue, pvalue = stat.f_oneway(data['Zone - A'], data['Zone - B'], data['Zone - C'],data['Zone - D'])
print('fvalue =',fvalue)
print('pvalue =',pvalue)
```

Since pvalue < 0.05 we reject null hypothesis.ie,all the zones generate different amount of sales.
