# Bar graph of male vs female buyers

```python
plt.figure(figsize=(6,4))
data.groupby('Buyer Gender')['Buyer Gender'].count().plot(kind='bar')
plt.title('Male vs Female Buyers')
```

# Top 5 cars based on their sales price

```python
data_top = data.nlargest(5, 'Sale Price')
```

```python
df=data_top[['Make','Model','Sale Price']]
```
